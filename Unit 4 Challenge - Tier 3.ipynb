**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

```python
# First step here will be group by year and get avg price (mean) by borough by year

properties_group = properties_melted.groupby(by = ['Borough', 'Year']).mean()
properties_group = properties_group.reset_index()
properties_group.sample(2)
```

```python
# Now we will want to find the ratio of price 2018 / price 1998
#  First we will get this for one borough


properties_Hackney = properties_group[properties_group['Borough'] == 'Hackney']

value_1998 = float(properties_Hackney['Average_Price'][properties_Hackney['Year']==1998])
value_2018 = float(properties_Hackney['Average_Price'][properties_Hackney['Year']==2018])
ratio_H = value_2018 / value_1998
ratio_H

# According to this analysis home prices in this borough have increased 6 fold over a 20 year period...seems too high 
# Next will attempt to use a similar analysis for all boroughs
```

```python
#  Generalizing previous analysis to incorporate into a loop
dict_ratio = {}
bor = 'Hackney'
value_1998 = float(properties_group['Average_Price'][properties_group['Borough'] == bor][properties_group['Year'] ==1998])
value_2018 = float(properties_group['Average_Price'][properties_group['Borough'] == bor][properties_group['Year'] ==2018])
ratio = value_2018 / value_1998
dict_ratio[bor] = ratio
print(dict_ratio)
```

```python
# Iterating through list of boroughs and adding ratio to a dictionary
borough_list = properties_melted['Borough'].unique()
dict_gen = {}
for bor in borough_list:
    value_1998 = float(properties_group['Average_Price'][properties_group['Borough'] == bor][properties_group['Year'] ==1998])
    value_2018 = float(properties_group['Average_Price'][properties_group['Borough'] == bor][properties_group['Year'] ==2018])
    ratio = value_2018 / value_1998
    dict_gen[bor] = [ratio]
print(dict_gen)
```

```python
# Convert dictionary list back to a dataframe
df = pd.DataFrame(dict_gen)
df
```

```python
#  Transpose and rename columns
df = df.transpose()
df= df.reset_index()
df.rename(columns = {'index': 'Borough', 0: 'Ratio'}, inplace = True)
df1 = df
```

```python
# Sort, reset index and print
df_sorted = df1.sort_values(by = 'Ratio', ascending = False)
df_sorted.reset_index()
df_sorted
```
